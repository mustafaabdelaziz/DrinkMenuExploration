```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('drinkMenu.csv')
data.head()
data.describe()
```

```python
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
plt.figure(figsize = (10, 6))
plt.bar(data['Beverage'], data['Calories'])
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.bar(data['Beverage'], data[' Sugars (g)'])
plt.xticks(rotation = 90)
plt.show()
```
